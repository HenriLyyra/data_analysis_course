```python
# IMPORTING

import numpy
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
# TASK 2.1

print("IMPORTING AND PRINTING THE ORIGINAL DATA MATRIX:")
data1=numpy.loadtxt("task_II_data.txt", dtype=float, delimiter=",")
print(data1)
print('\n' + "Length of the original list = " + str( len(data1) ) + '\n')

print("REMOVING ROWS WITH nan VALUES AND PRINTING THE EDITED DATA MATRIX:")
lData1 = len(data1);
for i in range( lData1 ):
    if numpy.isnan(data1[lData1 - 1 - i]).any():
        data1 = numpy.delete(data1, (lData1 - 1 - i), axis=0);
print(data1)
print('\n' + "Length of the edited list = " + str( len(data1) ) + '\n')

# TASK 2.2

print("PRINTING THE VECTOR CONTAINING MEANS OF THE COLUMNS OF THE EDITED DATA MATRIX:")
print( numpy.mean(data1, axis=0) )
print('\n' + "PRINTING THE VECTOR CONTAINING STANDARD DEVIATIONS OF THE COLUMNS OF THE EDITED DATA MATRIX: ")
print( numpy.std(data1, axis=0) )

# TASK 2.3

print('\n' + "The mean of the second row = " + str( numpy.mean(data1, axis=0)[1] ) )
print('\n' + "The standard deviation of the second row = " + str( numpy.std(data1, axis=0)[1] ) )

# TASK 2.4

print('\n' + "PRINTING THE THIRD ROW OF THE EDITED DATA MATRIX:")
print(data1[2])

print('\n' + "REPLACING THE THIRD ROW OF THE EDITED DATA MATRIX WITH ONES:")
data1[2] = numpy.ones(len(data1[2]))
print('\n' + "THE THIRD ROW OF THE EDITED DATA MATRIX IS NOW:")
print(data1[2])

# TASK 2.5

print('\n' + "PRINTING THE ROWS OF THE EDITED DATA MATRIX WHOSE SECOND ELEMENT IS GREATER OR EQUAL TO 0.5:")
for i in range( len(data1) ):
    if (data1[i,1] >= 0.5):
        print(data1[i])
```

```python
# TASK 3.1

print("IMPORTING TRANSPOSING AND PRINTING THE DATA MATRIX:")
data2 = numpy.transpose( numpy.loadtxt("task_III_data.txt", dtype=float, delimiter=",") )
print(data2)

# TASK 3.2

x = data2[0]
y = data2[1]

linData2 = interp1d(x, y, kind='linear')
cubData2 = interp1d(x, y, kind='cubic')

xnew = numpy.arange(x[0], x[len(x) - 1], 0.001 )
yLin = linData2(xnew)
yCub = cubData2(xnew)

plt.xlabel('Input value')
plt.ylabel('Function value')
plt.title('Input data and interpolated functions')
#plt.grid(False)

plt.plot(xnew, yLin, '-', label = 'linear')
plt.plot(xnew, yCub, '-', label = 'cubic')
plt.plot(x, y, 'ro', label = 'data')
plt.legend()

plt.show()
```

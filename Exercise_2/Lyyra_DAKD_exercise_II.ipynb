```python
import numpy as np;
import scipy.stats;
import math;
import matplotlib.pyplot as plt;
from mpl_toolkits.mplot3d import Axes3D 
```

```python
# TASK 1

print("IMPORTING THE IRIS DATA:")

# Iris data from  http://archive.ics.uci.edu/ml/datasets/Iris
#
# Attribute Information:
# 0. sepal length in cm
# 1. sepal width in cm
# 2. petal length in cm
# 3. petal width in cm
# 4. class/species: Iris Setosa, Iris Versicolour, Iris Virginica

irisData=np.loadtxt("iris.data", dtype=str, delimiter=",")
#print(irisData)

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("\nEXTRACTING THE DATA OF THE TOTAL SET:")

total = [];

for i in range( 0,len(irisData) ):
    total.append(irisData[i][0:4]);

total = np.array(total,dtype = float); 

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("\nEXTRACTING THE DATA OF SETOSA:")

setosa = [];

for i in range( 0,len(irisData) ):
    if 'Iris-setosa' in irisData[i]:
        setosa.append(irisData[i][0:4]);

setosa = np.array(setosa,dtype = float);        
        
#print(setosa)

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("\nEXTRACTING THE DATA OF VERSICOLOR:")

versicolor = [];

for i in range( 0,len(irisData) ):
    if 'Iris-versicolor' in irisData[i]:
        versicolor.append(irisData[i][0:4]);

versicolor = np.array(versicolor,dtype = float);        
        
#print(versicolor)

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("\nEXTRACTING THE DATA OF VIRGINICA:")

virginica = [];

for i in range( 0,len(irisData) ):
    if 'Iris-virginica' in irisData[i]:
        virginica.append(irisData[i][0:4]);

virginica = np.array(virginica,dtype = float);        
        
#print(virginica)
```

```python
# TASK 2

print("\nPRINTING THE VECTOR CONTAINING MEANS OF THE SETOSA ATTRIBUTES:")
print( np.mean(setosa, axis=0) )
print("\nPRINTING THE VECTOR CONTAINING STANDARD DEVIATIONS OF THE COLUMNS OF THE SETOSA ATTRIBUTES:")
print( np.std(setosa, axis=0) )

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("\nPRINTING THE VECTOR CONTAINING MEANS OF THE VERSICOLOR ATTRIBUTES:")
print( np.mean(versicolor, axis=0) )
print("\nPRINTING THE VECTOR CONTAINING STANDARD DEVIATIONS OF THE COLUMNS OF THE VERSICOLOR ATTRIBUTES:")
print( np.std(versicolor, axis=0) )

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("\nPRINTING THE VECTOR CONTAINING MEANS OF THE VIRGINICA ATTRIBUTES:")
print( np.mean(virginica, axis=0) )
print("\nPRINTING THE VECTOR CONTAINING STANDARD DEVIATIONS OF THE COLUMNS OF THE VIRGINICA ATTRIBUTES:")
print( np.std(virginica, axis=0) )
```

```python
# TASK 3

# BASED ON EXAMPLE FROM https://matplotlib.org/gallery/statistics/histogram_multihist.html?highlight=histogram

# FUNCTIONS TO DETERMINE THE NUMBER OF BINS
def binnum(h, data):
    return math.ceil( ( np.amax(data, axis=None) - np.amin(data, axis=None) ) / h )

def binSturges(data):
    return math.ceil( math.log( len(data), 2) + 1 )

def hScott(data):
    return 3.5*np.std(data, axis=0) / ( len(data) )**(1/3)

def binSqrt(data):
    return math.ceil( math.sqrt( len( data ) ) )

def hFreedmanDiaconis(data):
    return 2*scipy.stats.iqr(data,  rng=(25, 75) ) / ( len(data) )**(1/3)

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

# PLOTS ALL THE ATTRIBURTES OF data USING bins NUMBER OF BINS 
def histAttributes(data,bins):
    fig, axes = plt.subplots(nrows=2, ncols=2)
    att1, att2, att3, att4 = axes.flatten()

    att1.hist( np.transpose( data )[0], bins )
    att1.set_title('Attribute 1')
    
    att2.hist( np.transpose( data )[1], bins )
    att2.set_title('Attribute 2')
    
    att3.hist( np.transpose( data )[2], bins )
    att3.set_title('Attribute 3')
    
    att4.hist( np.transpose( data )[3], bins )
    att4.set_title('Attribute 4')
    

    fig.tight_layout()
    plt.show()

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

# PLOTS ALL THE ATTRIBURTES OF VARIABLES data1 and data1 USING bins NUMBER OF BINS 
def histCompAttributes(data1, data2, name1, name2, bins):
    fig, axes = plt.subplots(nrows=2, ncols=2)
    att1, att2, att3, att4 = axes.flatten()
    
    colors = ['orange', 'blue']
    names = [name1,name2]
    legendSize = 7
    
    x0 = [ np.transpose( data1 )[0], np.transpose( data2 )[0] ]

    att1.hist( x0, bins, histtype='bar', color=colors, label=names )
    att1.set_title('Attribute 1')
    att1.legend(prop={'size': legendSize})
    
    x1 = [ np.transpose( data1 )[1], np.transpose( data2 )[1] ]
    
    att2.hist( x1, bins, histtype='bar', color=colors, label=names )
    att2.set_title('Attribute 2')
    att2.legend(prop={'size': legendSize})
    
    x2 = [ np.transpose( data1 )[2], np.transpose( data2 )[2] ]
    
    att3.hist( x2, bins, histtype='bar', color=colors, label=names )
    att3.set_title('Attribute 3')
    att3.legend(prop={'size': legendSize})
    
    x3 = [ np.transpose( data1 )[3], np.transpose( data2 )[3] ]
    
    att4.hist( x3, bins, histtype='bar', color=colors, label=names )
    att4.set_title('Attribute 4')
    att4.legend(prop={'size': legendSize})
    

    fig.tight_layout()
    plt.show()
    
#histAttributes(setosa, binSqrt(setosa))
#histAttributes(versicolor, None)
#histAttributes(virginica, None)

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

# PLOTS nTH ATTRIBURTE OF data USING THE FOUR DIFFERENT BIN NUMBERS 
def histAttributeBins(data,n):# n GOES FROM 0 TO 3
    data2 = np.transpose( data )[n]
    
    fig, axes = plt.subplots(nrows=2, ncols=2)
    att1, att2, att3, att4 = axes.flatten()

    att1.hist( data2, binSturges(data2), color = 'green' )
    att1.set_title('Sturges')
    
    att2.hist( data2, binnum(hScott(data2), data2), color = 'green' )
    att2.set_title('Scott')
    
    att3.hist( data2, binSqrt(data2), color = 'green' )
    att3.set_title('Square root')
    
    att4.hist( data2, binnum(hFreedmanDiaconis(data2), data2), color = 'green' )
    att4.set_title('Freedman-Diaconis')
    

    fig.tight_layout()
    plt.show()

#for attn in range(0,4):
#
#    print("\nSHOWING THE PLOTS OF ATTRIBUTE " + str(attn) + " OF SETOSA:")
#    histAttributeBins(setosa,attn)
    
print("\nSHOWING THE PLOTS OF ATTRIBUTE " + str(2) + " OF SETOSA:")
histAttributeBins(setosa,2)
print("\nWe notice, that Sturges seems to produce the narrowest distribution for this data. Scott shows a a distribution heavier on the right-hand side. Square root and Freedman-Diaconis seem quite similar to each other.")
    
## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 
for nbins in range(5,17):
    print("\nSHOWING THE ATTRIBUTES OF VIRGINICA AND VERSICOLOR USING " + str(nbins) + " BINS:")

    histCompAttributes(versicolor, virginica, 'Versicolor', 'Virginica', nbins)

print("We notice that the two species can be best distinguished by the petal width and petal length.")
```

```python
# TASK 5

print("MAKING SCATTER PLOT TABLE OF ATTRIBUTES FOR EACH SPECIES:")

fig = plt.figure(figsize=(15,15))
fig.suptitle('Scatter plots of attributes w.r.t. each other:', fontsize=36)

position = 0
for i in range(0,4):
    for j in range(0,4):
        position = position + 1
        if i == j:
            if i == 0:
                string = "Sepal length in cm"
            if i == 1:
                string = "Sepal width in cm"
            if i == 2:
                string = "Petal length in cm"
            if i == 3:
                string = "Petal width in cm"
            plt.subplot( 4, 4, position )
            plt.text(0.25, 0.5, string )
        else:
            plt.subplot( 4, 4, position )
            plt.scatter( np.transpose( setosa )[i], np.transpose( setosa )[j], color='green' )
            plt.scatter( np.transpose( versicolor )[i], np.transpose( versicolor )[j], color='orange' )
            plt.scatter( np.transpose( virginica )[i], np.transpose( virginica )[j], color='blue' )
plt.show()

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

# VISUALIZING THE CORRELATION COEFFICIENTS, BY FOLLOWING THE EXAMPLE FROM https://matplotlib.org/gallery/mplot3d/3d_bars.html

def pearson( species ):
    return np.corrcoef(np.transpose( species ), y=None, rowvar=True)

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

def spearmanRho( species ):
    return scipy.stats.spearmanr( species  )[0]

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

def kendallTau( species ):

    mat = np.zeros( (4,4) )

    for i in range(0,4):
        for j in range(0,4):
            mat[i,j] = scipy.stats.kendalltau( np.transpose( species )[i], np.transpose( species )[j] )[0]
    return mat

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

def printDataBars( speciesData0, speciesData1, speciesData2, speciesData3, title ):
    # setup the figure and axes
    fig = plt.figure(figsize=(20,20))
    fig.suptitle( title, fontsize=36 )

    ax0 = fig.add_subplot(2,2,1, projection='3d')
    ax1 = fig.add_subplot(2,2,2, projection='3d')
    ax2 = fig.add_subplot(2,2,3, projection='3d')
    ax3 = fig.add_subplot(2,2,4, projection='3d')

    # coordinates (attribute number)
    _x = np.arange(4)
    _y = np.arange(4)
    _xx, _yy = np.meshgrid(_x, _y)
    x, y = _xx.ravel(), _yy.ravel()

    bottom = np.zeros(16)
    width = depth = 0.3

    colors = ['grey','green','orange','blue']

    # VALUE DEPENDENT COLORS, FOLLOWING THE EXAMPLE FROM https://stackoverflow.com/questions/43869751/change-bar-color-in-a-3d-bar-plot-in-matplotlib-based-on-value
    data0 = speciesData0.flatten()
    colors0 = plt.cm.jet( np.sign(data0) )

    top0 = np.abs( data0 )

    top1 = speciesData1.flatten()

    top2 = speciesData2.flatten()

    top3 = speciesData3.flatten()

    angle = -25-90

    ax0.view_init(30, angle)
    ax0.bar3d(x, y, bottom, width, depth, top0,shade=True, color = colors0 )#colors[0])
    ax0.set_xlabel('Attribute', fontsize = 26)
    ax0.set_ylabel('Attribute', fontsize = 26)
    ax0.set_zlabel('Correlation', fontsize = 26)
    ax0.set_title('Total\nblue bar = negative value', fontsize = 26)

    ax1.view_init(30, angle)
    ax1.bar3d(x, y, bottom, width, depth, top1,shade=True, color = colors[1])
    ax1.set_xlabel('Attribute', fontsize = 26)
    ax1.set_ylabel('Attribute', fontsize = 26)
    ax1.set_zlabel('Correlation', fontsize = 26)
    ax1.set_title('Setosa', fontsize = 26)

    ax2.view_init(30, angle)
    ax2.bar3d(x, y, bottom, width, depth, top2,shade=True, color = colors[2])
    ax2.set_xlabel('Attribute', fontsize = 26)
    ax2.set_ylabel('Attribute', fontsize = 26)
    ax2.set_zlabel('Correlation', fontsize = 26)
    ax2.set_title('Versicolor', fontsize = 26)

    ax3.view_init(30, angle)
    ax3.bar3d(x, y, bottom, width, depth, top3,shade=True, color = colors[3])
    ax3.set_xlabel('Attribute', fontsize = 26)
    ax3.set_ylabel('Attribute', fontsize = 26)
    ax3.set_zlabel('Correlation', fontsize = 26)
    ax3.set_title('Virginica', fontsize = 26)

    plt.show()


## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("\nPRINTING THE PEARSON'S CORRELATION COEFFICIENT TABLES OF ATTRIBUTES:")

print("\nTotal:")
print( pearson(total) )

print("\nSetosa:")
print( pearson( setosa ) )

print("\nVersicolor:")
print( pearson( versicolor ) )

print("\nVirginica:")
print( pearson( virginica ) )
    
printDataBars( pearson( total ), pearson( setosa ), pearson( versicolor ), pearson( virginica ), 'Pearson\'s coefficient tables of attributes:' )    

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

print("NOTE: here we consider values 0.8 and higher Pearson\'s correlation coefficient as linearly correlated:")

print("In total data, we see that there is linear correlation between the attribute pairs (sepal length, petal length), (sepal length, petal width) and (petal length, petal width).")
print("Also, there seem to be small, yet significant, negative linear correlations between (sepal width, petal length) and (sepal width, petal width)")
print("\nFor Setosa and Versicolor, we see no high linear correlations, but neither negative linear correlations.")
print("\nFor Virginica, we see that sepal length and sepal width are linearly correlated, and there are no negative linear correlations between any attributes.")
print("\nCONCLUSIONS: \nWe note, that in the data of individual species, all correlation coefficients are positive for all attribute pairs. Contrary to this, when looking at the total Iris data, we see also negative correlation coefficients.")

print("\nSince Pearson\'s correlation coefficient quantifies only linear correlations, the above does not exclude other, such as non-linear monotonic correlations for any of the attributes.")

print("\n-- -- -- -- -- -- -- -- -- -- -- -- -- -- ")

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

#EXPLANATION OF SRPEARMAN'S RHO FROM https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient
print("\nSpearman\'s rho is a nonparametric measure of rank correlation (statistical dependence between the rankings of two variables). It assesses how well the relationship between two variables can be described using a monotonic function.")

print("\nPRINTING THE SPEARMAN'S RHO CORRELATION COEFFICIENT TABLES OF ATTRIBUTES:")

print("\nTotal:")
print( spearmanRho(total) )

print("\nSetosa:")
print( spearmanRho(setosa) )

print("\nVersicolor:")
print( spearmanRho(versicolor) )

print("\nVirginica:")
print( spearmanRho(virginica) )

printDataBars( spearmanRho( total ), spearmanRho( setosa ), spearmanRho( versicolor ), spearmanRho( virginica ), 'Spearman\'s rho tables of attributes:' )    

print("\n-- -- -- -- -- -- -- -- -- -- -- -- -- -- ")

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 

# EXPLANATION OF KENDALL'S TAU FROM https://en.wikipedia.org/wiki/Kendall_rank_correlation_coefficient
print("\nKendall\'s tau is a measure the ordinal association between two measured quantities")
print("\nKendall\'s tau between two attributes will be high when observations have a similar rank (i.e. relative position label of the observations within the attribute: 1st, 2nd, 3rd, etc.) between the two attributes, and low when observations have a dissimilar rank between the two attributes.")
      
print("\nPRINTING THE KENDALL'S TAU CORRELATION COEFFICIENT TABLES OF ATTRIBUTES:")


print("\nTotal:")
print( kendallTau(total) )

print("\nSetosa:")
print( kendallTau(setosa) )

print("\nVersicolor:")
print( kendallTau(versicolor) )

print("\nVirginica:")
print( kendallTau(virginica) )

printDataBars( kendallTau( total ), kendallTau( setosa ), kendallTau( versicolor ), kendallTau( virginica ), 'Kendall\'s tau tables of attributes:' )

## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## ## 
```

```python
# TASK 6

# METHOD FOR CALCULATING THE PRINCIPAL COMPONENT ANALYSIS COPIED FROM THE LECTURE SLIDES, LECURE 4

def pCA( species ):

    m = species.shape[0]
    Z = species - np.mean(species, axis=0)
    C = Z.T@Z/(m-1)
    evals, evecs = np.linalg.eigh(C)
    # sort eigenvalues in decreasing order
    idx = np.argsort(evals)[::-1]
    evecs = evecs[:,idx]
    #pick eigenvectors corresponding to 2 largest eigenvalues
    M = evecs[:,:2].T
    Z_mapped = (M@Z.T).T
    
    return Z_mapped.T


# METHOD FOR CALCULATING THE PRINCIPAL COMPONENT ANALYSIS WITH Z-SCORE NORMALIZATION

def pCAZScoreNorm( species ):

    m = species.shape[0]
    Z = (species - np.mean(species, axis=0))/np.std(species, axis=0)
    C = Z.T@Z/(m-1)
    evals, evecs = np.linalg.eigh(C)
    # sort eigenvalues in decreasing order
    idx = np.argsort(evals)[::-1]
    evecs = evecs[:,idx]
    #pick eigenvectors corresponding to 2 largest eigenvalues
    M = evecs[:,:2].T
    Z_mapped = (M@Z.T).T
    
    return Z_mapped.T
    
    #plt.scatter( Z_mapped.T[0], Z_mapped.T[1] )
    #plt.show()
    
legendSize = 12
    
plt.scatter( pCA( total )[0][0:len(setosa)], pCA( total )[1][0:len(setosa)], color='green', label = 'Setosa' )
plt.scatter( pCA( total )[0][len(setosa) + 1:len(setosa) + 1 + len(versicolor)], pCA( total )[1][len(setosa) + 1:len(setosa) + 1 + len(versicolor)], color='orange', label = 'Versicolor' )
plt.scatter( pCA( total )[0][len(setosa) + 1 + len(versicolor):len(total)], pCA( total )[1][len(setosa) + 1 + len(versicolor):len(total)], color='blue', label = 'Virginica' )
plt.suptitle('Scatter plot of principal component analysis for the total data:\nNo z-score normalization', fontsize=14)
plt.legend(prop={'size': legendSize})
plt.show()

print("The variances of the primary components are " + str( np.var(pCA( total )[0] ) ) + " and " + str( np.var(pCA( total )[1] )  ) )
print("The ratio of the variances of the primary components is " + str( np.var(pCA( total )[0] / np.var(pCA( total )[1] )  ) ) )
    
plt.scatter( pCAZScoreNorm( total )[0][0:len(setosa)], pCAZScoreNorm( total )[1][0:len(setosa)], color='green', label = 'Setosa' )
plt.scatter( pCAZScoreNorm( total )[0][len(setosa) + 1:len(setosa) + 1 + len(versicolor)], pCAZScoreNorm( total )[1][len(setosa) + 1:len(setosa) + 1 + len(versicolor)], color='orange', label = 'Versicolor' )
plt.scatter( pCAZScoreNorm( total )[0][len(setosa) + 1 + len(versicolor):len(total)], pCAZScoreNorm( total )[1][len(setosa) + 1 + len(versicolor):len(total)], color='blue', label = 'Virginica' )
plt.suptitle('Scatter plot of principal component analysis for the total data:\nWith z-score normalization', fontsize=14)
plt.legend(prop={'size': legendSize})
plt.show()

print("The variances of the primary components are " + str( np.var(pCAZScoreNorm( total )[0] ) ) + " and " + str( np.var(pCAZScoreNorm( total )[1] )  ) )
print("The ratio of the variances of the primary components is " + str( np.var(pCAZScoreNorm( total )[0] / np.var(pCAZScoreNorm( total )[1] )  ) ) )
 
print("\nWe observe, that Setosa can be easily distinguished from other species, and z-score normalization decreases the variance of the first primary component while it increases the variance of the second one.")

# BELOW: PCA FOR INDIVIDUAL SPECIES
# CONCLUSION: NOT INTERESTING

#plt.scatter( pCA( setosa )[0], pCA( setosa )[1], color='green', label = 'Setosa' )
#plt.suptitle('Scatter plot of principal component analysis for the setosa data:', fontsize=14)
#plt.legend(prop={'size': legendSize})
#plt.show()

#plt.scatter( pCA( versicolor )[0], pCA( versicolor )[1], color='orange', label = 'Versicolor' )
#plt.suptitle('Scatter plot of principal component analysis for the versicolor data:', fontsize=14)
#plt.legend(prop={'size': legendSize})
#plt.show()

#plt.scatter( pCA( virginica )[0], pCA( virginica )[1], color='blue', label = 'Virginica' ) 
#plt.suptitle('Scatter plot of principal component analysis for the virginica data:', fontsize=14)
#plt.legend(prop={'size': legendSize})
#plt.show()
```
